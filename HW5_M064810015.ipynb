# 載入keras

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout

import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
submit = pd.read_csv('submission.csv')
```

# 選取特徵項目

```python
selected_features  = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex', 'Embarked']
```

```python
X_train = train[selected_features]
```

```python
y_train = train['Survived']
```

```python
X_test = test[selected_features]
```

# 處理缺失值 

```python
X_train.info()
```

```python
X_test.info()
```

```python
X_train['Age'].fillna(X_train['Age'].mean(), inplace=True)
X_train['Embarked'].fillna('S', inplace=True)
X_test['Age'].fillna(X_test['Age'].mean(), inplace=True)
X_test['Fare'].fillna(X_test['Fare'].mean(), inplace=True)
```

```python
X_train.info()
```

# 將字串類型的類別資料做轉換

```python
X_train = pd.get_dummies(X_train)
```

```python
X_train.head()
```

```python
X_test = pd.get_dummies(X_test)
```

```python
X_test.head()
```

# 建立keras模型

```python
model = Sequential()
```

```python
model = Sequential()  
model.add(Dense(units=50,input_dim = 10,kernel_initializer='uniform',activation='relu'))  
model.add(Dense(30,activation='relu'))  
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
train_his = model.fit(X_train ,y_train ,validation_split=0.1, epochs=100, batch_size=32, verbose=2)
```

```python
scores = model.evaluate(X_train, y_train)
```

# 預測

```python
survived_predict = model.predict(X_test)
survived_predict = np.array(survived_predict)
survived_predict = np.round(survived_predict)
survived_predict
```

# 輸出

```python
submission=pd.DataFrame({'PassengerId':test['PassengerId'], 'Survived':survived_predict[0:418,0]})
```

```python
submission.to_csv('submission.csv', index=False)
```
